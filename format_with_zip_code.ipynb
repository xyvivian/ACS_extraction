```python
acs_data = pd.read_csv("result.csv")
acs_data
```

```python
acs_data.drop('Block Group',inplace=True, axis = 1)
acs_data
```

```python
censusdata.printtable(censusdata.censustable('acs5', 2018, 'B08301'))
```

```python
def find_zipcode(block_names):
    val = block_names.params()
    code = int(val[0][1] + val[1][1] + val[2][1])
    ind = zip_code.index[zip_code['tract'] == code]
    zipcode = zip_code.loc[ind,'zip']
    zp =  zipcode.to_string().split()[1]
    return zp
```

```python
vehicle_data = censusdata.download('acs5', 2018,
           censusdata.censusgeo([('state', '12'),
                                 ('county', '091'),
                                 ('block group', '*')]),['B08301_002E','B08301_010E','B08301_017E'], key= key)

for i in county_list[1:]:
    vehicle_data_new = censusdata.download('acs5', 2018,
           censusdata.censusgeo([('state', '12'),
                                 ('county', i),
                                 ('block group', '*')]),['B08301_002E','B08301_010E','B08301_017E'], key=key)
    vehicle_data = vehicle_data.append(vehicle_data_new)


vehicle_data['Transportation: Cars'] = vehicle_data.B08301_002E
vehicle_data['Transportation: Public'] = vehicle_data.B08301_010E
vehicle_data['Transportation: Motor'] = vehicle_data.B08301_017E
vehicle_data = vehicle_data[['Transportation: Cars', 'Transportation: Public', 'Transportation: Motor']]
vehicle_data
```

```python
block_group_names = vehicle_data.index
acs_data.index = block_group_names
acs_data
```

```python
acs_data_new = pd.concat([acs_data,vehicle_data],axis=1,sort=False)
acs_data_new
```

```python
zp_list = []
for i in list(acs_data_new.index.to_list()):
    zp = find_zipcode(i)
    zp_list.append(zp)
    
acs_data_new.insert(0, "Zip Code", zp_list, True)
acs_data_new
```

```python
zp_unique_values = list(set(zp_list))
zp_unique_values
```

```python
res = acs_data_new.groupby('Zip Code', as_index=True).sum()
res
```
